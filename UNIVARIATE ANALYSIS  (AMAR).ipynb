# STATS

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("Book_ipl22_ver_33.csv")


print("\nSummary Statistics for Numerical Columns:")
display(df.describe())


print("\nCategorical Summary:")
display(df.describe(include=["object"]))

sns.set(style="whitegrid")
```

# HISTOGRAM & AREA CHART

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

numerical_cols = df.select_dtypes(include=['number']).columns

for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], bins=15, kde=True, color="blue")
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python

plt.figure(figsize=(12, 6))
df[numerical_cols].plot(kind="area", alpha=0.4, figsize=(12, 6))
plt.title("Area Chart for Numerical Features")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend(loc="upper right")
plt.show()
```

# BAR CHART & LINE CHART

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(style="whitegrid")


categorical_cols = ["venue", "match_winner", "toss_decision", "player_of_the_match"]


for col in categorical_cols:
    plt.figure(figsize=(10, 4))
    sns.countplot(y=df[col], order=df[col].value_counts().index, palette="viridis")
    plt.title(f"Bar Chart of {col}")
    plt.xlabel("Count")
    plt.ylabel(col)
    plt.show()
```
